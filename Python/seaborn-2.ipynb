```python
# import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Box Plot

```python
# # sample syntax
# seaborn.boxplot(x='',y='',data=f)
# plt.show()
```

```python
d1 = pd.read_csv ('/kaggle/input/tipping/tips.csv')
```

```python
d1.head()
```

```python
sns.set(style='whitegrid')
```

```python
sns.boxplot(x='day',y='total_bill',data=d1)
plt.show()
```

```python
# diff into time 
sns.boxplot(x='day',y='total_bill',data=d1,
           hue = 'time')
plt.show()
```

```python
# change in order 
sns.boxplot(x='day',y='total_bill',data=d1,
           hue = 'sex',
           order = ['Sat','Sun','Fri','Thur'])
plt.show()
```

```python
# show means 
# change in order 
sns.boxplot(x='day',y='total_bill',data=d1,
           hue = 'sex',
           order = ['Sat','Sun','Fri','Thur'],
           showmeans= True)
plt.show()
```

```python
# change in means properties
# show means 
# change in order 
sns.boxplot(x='day',y='total_bill',data=d1,
           hue = 'sex',
           order = ['Sat','Sun','Fri','Thur'],
           showmeans= True,
           meanprops={'marker':'*','markeredgecolor':'r'},
           )
plt.show()
```

```python
# change in line width
# change in means properties
# show means 
# change in order 
sns.boxplot(x='day',y='total_bill',data=d1,
           hue = 'sex',
           order = ['Sat','Sun','Fri','Thur'],
           showmeans= True,
           meanprops={'marker':'*','markeredgecolor':'r'},
            linewidth=3,
            
           )
plt.show()
```

```python
# change in color
sns.boxplot(x='day',y='total_bill',data=d1,
           hue = 'sex',
           order = ['Sat','Sun','Fri','Thur'],
           showmeans= True,
           meanprops={'marker':'*','markeredgecolor':'r'},
            linewidth=3,
            palette = 'Accent',
            
            
           )
plt.show()
```

```python
#horizontal
# change in color
sns.boxplot(y='day',x='total_bill',data=d1,
           hue = 'sex',
           order = ['Sat','Sun','Fri','Thur'],
           showmeans= True,
           meanprops={'marker':'*','markeredgecolor':'r'},
            linewidth=3,
            palette = 'Accent',
            # orient= 'h' # numerical only 
            
            
           )
plt.show()
```

```python
# single box 
 
sns.boxplot(y=d1['total_bill']
            
           )
plt.show()
```

## How to change styling in plot

* change in figure style
* change in Axes spines
* change in scale and contexr

```python
d1.head(0)
```

```python


sns.barplot(x='day',y='total_bill',data=d1)
plt.show()
```

```python
 # change in figure sizre

plt.figure(figsize=(12,12))
sns.barplot(x='day',y='total_bill',data=d1)
sns.despine()
plt.show()
```

```python
 # change in figure sizre

plt.figure(figsize=(12,12))

sns.set_context("paper",font_scale=1)
sns.barplot(x='day',y='total_bill',data=d1)
sns.despine()
plt.show()
```
