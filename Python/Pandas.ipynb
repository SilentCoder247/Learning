# Merge 

```python
import pandas as pd
```

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':['a','b','c','d']})
d2= pd.DataFrame({'A':[1,2,3,5],'C':['aa','bb','cc','dd']})

d1 
```

```python
d2
```

```python
s3=pd.merge(d1,d2,on='A')
s3
```

```python
s3=pd.merge(d1,d2,how='inner')
s3
```

```python
s3=pd.merge(d1,d2,how='left')
s3
```

```python
s3=pd.merge(d1,d2,how='right')
s3
```

```python
s3=pd.merge(d1,d2,how='outer')
s3
```

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':['a','b','c','d']})
d2= pd.DataFrame({'A':[1,2,3,5],'B':['aa','bb','cc','dd']})

d1 
```

```python
d2
```

# Join

```python

d1= pd.DataFrame({'A':[1,2,3,5],'B':['aa','bb','cc','dd']})

d2= pd.DataFrame({'C':[11,12,15],'D':['a','b','c']})
```

```python
 d1.join(d2)
```

```python
 d1.join(d2,how='right')
```

```python
 d1.join(d2,how='left')
```

```python
 d1.join(d2,how='outer')
```

```python

d1= pd.DataFrame({'A':[1,2,3,5],'B':['aa','bb','cc','dd']})

d2= pd.DataFrame({'C':[11,12,15],'B':['a','b','c']})
d1.join(d2,how='inner')
```

```python

d1= pd.DataFrame({'A':[1,2,3,5],'B':['aa','bb','cc','dd']})

d2= pd.DataFrame({'C':[11,12,15],'B':['a','b','c']})
d1.join(d2,how='inner',lsuffix='_12')
```
