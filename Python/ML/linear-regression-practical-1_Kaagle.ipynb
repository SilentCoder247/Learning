```python
# train test split 
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state = 42)
```

```python
# apply Linear regration  
```

```python
## fisrt make standardiation on dataset 

# we alway apply fit transform ---- x_trai 
# we alway apply tranform  ---- x_test




### but but We never apply tranform on out put variable
```

```python
# standardizing the dataset
```

```python
 from sklearn.preprocessing import StandardScaler
```

```python
# initilization of standard  SCALER 
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
# now we standar dize our data 
```

```python
# let's make our model now 
from sklearn.linear_model import LinearRegression
```

```python
# cross validation  
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
mse = cross_val_score (regression,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
np.mean(mse)
```

```python
# Prediction 
```

```python
reg_Pred=regression.predict(X_test)
```

```python
reg_Pred
```

```python
import seaborn as sns
```

```python
sns.displot(reg_Pred-y_test,kind='kde')
```

```python
from sklearn.metrics import r2_score
```
