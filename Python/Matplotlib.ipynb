```python
# make two bar graph in same graph  side by side
import matplotlib.pyplot as plt
import numpy as np
x = ['c','c++','python','java']
y = [125,26,48,95]
z= [10,20,30,50]

width = 0.3
p=np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel('language',fontsize = 20)
plt.ylabel('student',fontsize = 20)
plt.title('Krunal',fontsize = 20)

c=['y','b','m','g']
c1 =['r','m','g','y']

plt.bar(p,y,width,color=c,label='matplotlib')
plt.bar(p1,z,width,color=c1,label='mat')

# plt.xticks(p+width/2,x)

plt.legend()
plt.show ()
```

# Fill_Between Plot

```python
import matplotlib.pyplot as plt
x =[1,2,3,4,5,6]
y= [1,2,3,4,5,6]

plt.plot (x,y,color='r')

# plt.fill_between(x,y)
plt.fill_between(x,y,
                 color='y',
                where=(x>2) & (x<5))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x =np.array([1,2,3,4,5,6])
y= np.array([1,2,3,4,5,6])

plt.plot (x,y,color='r')

# plt.fill_between(x,y)
plt.fill_between(x,y,
                 color='y',
                where=(x>2) & (x<5),
                alpha =0.5)

plt.show()
```
